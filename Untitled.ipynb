```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, roc_curve, auc
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import pickle

# Load the data
data = pd.read_csv('diabetes.csv')
df = pd.DataFrame(data)

# Identify and remove outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Separate features and target
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=2)

# Hyperparameter tuning using RandomizedSearchCV
param_grid_rf = {'n_estimators': [100, 200, 300], 'max_depth': [None, 10, 20, 30]}
param_grid_lr = {'C': [0.01, 0.1, 1, 10, 100]}
param_grid_svm = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']}
param_grid_knn = {'n_neighbors': [3, 5, 7, 9]}

models = {
    'RandomForest': (RandomForestClassifier(random_state=42, class_weight='balanced'), param_grid_rf),
    'LogisticRegression': (LogisticRegression(random_state=42, class_weight='balanced'), param_grid_lr),
    'SVM': (SVC(random_state=42, class_weight='balanced', probability=True), param_grid_svm),
    'KNN': (KNeighborsClassifier(), param_grid_knn)
}

best_model = None
best_score = 0
model_scores = []
model_roc_curves = {}
model_reports = []

for name, (model, param_grid) in models.items():
    grid_search = RandomizedSearchCV(model, param_grid, cv=5, n_iter=10, random_state=42, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    y_pred = grid_search.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    model_scores.append((name, score))
    print(f'{name} Accuracy: {score} with best parameters: {grid_search.best_params_}')
    
    report = classification_report(y_test, y_pred, output_dict=True)
    model_reports.append(pd.DataFrame(report).transpose())
    print(classification_report(y_test, y_pred))
    
    # Plot confusion matrix
    disp = ConfusionMatrixDisplay.from_estimator(grid_search, X_test, y_test)
    disp.ax_.set_title(f'{name} Confusion Matrix')
    plt.show()
    
    # ROC curve data
    if hasattr(grid_search, "predict_proba"):
        y_score = grid_search.predict_proba(X_test)[:, 1]
    else:  # for SVM
        y_score = grid_search.decision_function(X_test)
    
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    model_roc_curves[name] = (fpr, tpr, roc_auc)
    
    if score > best_score:
        best_score = score
        best_model = grid_search.best_estimator_

# Save the best model
with open('best_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

# Save the scaler
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print(f'Best Model: {best_model} with accuracy: {best_score}')

# Combine model reports into a single DataFrame
comparison_df = pd.concat(model_reports, keys=models.keys())

# Display the comparison table
print(comparison_df)

# Plot accuracy comparison
model_names, scores = zip(*model_scores)
plt.figure(figsize=(10, 6))
plt.bar(model_names, scores, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.show()

# Plot ROC curves comparison
plt.figure(figsize=(10, 8))
for name, (fpr, tpr, roc_auc) in model_roc_curves.items():
    plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc="lower right")
plt.show()

# Feature importance plot for Random Forest
if isinstance(best_model, RandomForestClassifier):
    feature_importances = best_model.feature_importances_
    feature_names = df.columns[:-1]
    sorted_idx = np.argsort(feature_importances)
    
    plt.figure(figsize=(10, 8))
    plt.barh(range(len(sorted_idx)), feature_importances[sorted_idx], align='center')
    plt.yticks(range(len(sorted_idx)), [feature_names[i] for i in sorted_idx])
    plt.xlabel('Feature Importance')
    plt.title('Feature Importance for Random Forest')
    plt.show()
```
